## 1) Pandas

O pandas é uma das bibliotecas mais usadas em data science.

Seu objetivo é a **leitura, processamento e manipulação de dados** na forma de tabelas chamdas de **"DataFrame"**

Antes de conhecer o Pandas, vamos ler o arquivo csv "alunos.csv", da forma como aprendemos na aula de arquivos

````python
import csv

f = open("../datasets/alunos.csv", "r", encoding="utf-8")

leitor = csv.reader(f, delimiter=';', lineterminator='\n')

planilha = []

for linha in leitor:
    
    planilha.append(linha)
    
f.close()

planilha
````

Como fizemos na aula, uma vez lido o arquivo, é possível processá-lo de diversas maneiras.

Por exemplo, para obter **a primeira coluna**, isto é, os nomes, fazemos:

````python
[item[1] for item in planilha]
````

Agora, vamos usar o Pandas e aprender uma forma muito mais natural de processar dados!

````python
# importando a biblioteca
import pandas as pd
````

````python
# lendo o arquivo

tabela = pd.read_csv("../datasets/alunos.csv", sep=";")

tabela
````

Podemos **acessar os valores nas colunas** pelo nome delas:

````python
nome_da_coluna = "RA"

tabela[nome_da_coluna]
````

Dá pra **selecionar apenas algumas colunas** do dataframe (ou seja, criando um sub-dataframe):

````python
# pegando apenas a colua de nome e de média
tabela[["Nome", "Prova_1"]]
````

É possível **alterar valores** da tabela. Para isso, primeiro localizamos o valor a ser alterado com o **.loc**, passando a linha e coluna correspondente, e depois atribuímos o novo valor

Também é possível **criar uma linha nova**:

Podemos fazer operações entre os valores das colunas, e criar com isso novas colunas!

````python
# calculando a média usando as colunas

tabela["média"] = (tabela["Prova_1"] + tabela["Prova_2"] + 
                   tabela["Prova_3"] + tabela["Prova_4"])/4
````

````python
tabela
````

Também há algumas funções prontas:

````python
# calculando a media com a função mean()

tabela["media_2"] = tabela[["Prova_1", "Prova_2", "Prova_3", "Prova_4"]].mean(axis=1)

tabela
````

Naturalmente, o resultado é o mesmo!

Podemos **fazer filtros** muito facilmente

Basta explicitarmos **condições sobre os valores das colunas**, e utilizar isso como indexador do dataframe!

````python
# retorna o sub-dataframe que contém valores maiores que 7 na coluna "média"
# ou seja, é um filtro que utiliza a coluna "média"!

# dica: fazer o filtro no excel para ajudar na visualização

tabela[tabela["média"] > 7]
````

Se quisermos fazer filtros mais complexos (filtros compostos, em mais de uma coluna), podemos fazer **conjunções entre filtros**, utilizando os **operadores lógicos de conjunção**.

Obs.: temos os seguintes operadores lógicos:

- &     - corresponde ao "and"
- |     - corresponde ao "or"
- ~     - corresponde ao "not"

````python
tabela[(tabela["média"] > 7) & (tabela["Frequencia"] >= 20)]
````

````python
tabela[(tabela["média"] > 7) | (tabela["Frequencia"] >= 20)]
````

````python
# pegando as médias dos alunos que tiveram nota igual na prova 1 e na prova 4

tabela[tabela["Prova_4"] == tabela["Prova_1"]]["média"]
````

````python
tabela[tabela["Prova_4"] == tabela["Prova_1"]]["média"].mean()
````

````python
tabela[tabela["Prova_4"] < tabela["Prova_1"]]["média"].mean()
````

````python
tabela[tabela["Prova_4"] > tabela["Prova_1"]]["média"].mean()
````

````python
tabela[(tabela["Prova_4"] > tabela["Prova_1"])]
````

Agora que entendemos na prática como usar o Pandas, vamos nos aprofundar um pouco mais em sua estrutura!

O objeto fundamental do Pandas são as **Series**.

As Series são as **colunas das tabelas**, que funcionam como uma espécie de lista.

A diferença é que a série possui um **índice associado**, permitindo o acesso aos conteúdos dessa estrutura por ele, como um dicionário.


Podemos criar uma série **a partir de uma lista**, usando a função do pandas `pd.Series()`: 

````python
# definindo uma série com valores e indices

indices = ["a", "b", "c", "d", "e"]
lista = [10, 20, 30, 40, "aaaaa"]

serie = pd.Series(data = lista, index = indices)

serie
````

Séries de fato são parecidas com dicionários!

Com isso, podemos transformá-la em dicionários:

````python
# trasnformando a série em dicionario

dicionario = dict(serie)

dicionario
````

Assim, também podemos **criar uma série a partir de um dicionário**, e os índices e valores são automaticamente capturados:


````python
# criando uma série a partir de um dicionario

dic2 = {"nome": "André", 
        "idade" : 23}

pd.Series(dic2)
````

Agora que entendemos a componente fundamental do Pandas, as Séries, vamos falar um pouco mais sobre o **DataFrame**

O DataFrame é a estrutura que se assemelha a uma tabela/planilha, como vimos acima.

Ao ler um arquivo, um dataframe é criado automaticamente pelo pandas!

Por debaixo dos panos, o dataframe é representado por um dicionário em que a **chave** é o **nome da coluna** e os **valores** são as **Series** (todas com mesmo índice).

Assim, podemos **criar um dataframe a partir de um dicionario**, usando a função `pd.DataFrame()` 

````python
cadastro = {"nomes" : ["Joãozinho", "Mariazinha"],
                "idade" : [22, 25],
                "cidade" : ["Mauá", "Santo André"],
                "filhos": [0, 0],
                "altura" : [1.80, 1.65]}

cadastro
````

````python
# criando um dataframe a partir de um dicionario

df = pd.DataFrame(cadastro)
df
````

Agora que temos uma base mais complexa, vamos aproveitar para ver agora algumas outras funcionalidades do Pandas!

É possível utilizar o pandas para fazer uma **tabela dinâmica**:

````python
# tabela dinâmica agrupada por duas classes
# aqui, usamos a FUNÇÃO DE AGREGAÇÃO mean()

df.groupby(["Pclass", "Sex"]).mean()
````

Uma grande funcionalidade do pandas é o método `apply()`, utilizado para aplicar uma **função** (muitas vezes, uma **função lambda**) a uma coluna ou linha de um DataFrame

Uma função lambda nada mais é que uma **forma alternativa de declarar uma função**, de um jeito mais direto

````python
def dobro(x):
    
    return 2*x

dobro(2)
````

````python
faz_dobro = lambda x: 4*x
````

````python
faz_dobro(6)
````

Vamos selecionar a coluna de idades...

````python
df["Age"]
````

Aplicando uma função lambda **a todos os elementos da coluna**, ou seja, **à todas as linhas da tabela, daquela coluna específica**:

Tomando cada idade + 2, usando a função lambda definida.

Essa função lambda é equivalente a:

```python

def funcao(x)

    return x + 2
```

````python
df["Age"].apply(lambda x: x + 2)
````

Um outro exemplo:

````python
# função: transforma todos os números em string, e concatena "!!!!!!!!!" à string

df["Age"].apply(lambda x: str(x) + "!!!!!!!!!")
````

Vamos usar uma função lambda para **extrair o sobrenome** dos nomes dos passageiros

Pra extrarir o sobrenome, note que este está separada do resto do nome por vírgula.

Para perceber isso, dê uma olhada na coluna de nomes:

````python
df["Name"]
````

Portanto, podemos usar a função para strings `split(",")`, com quebra na vírgula, e depois selecionar o primeiro elemento da lista gerada!

Vamos aproveitar e **criar uma nova coluna da base**, com os sobrenomes!

````python
df["Surname"] = df["Name"].apply(lambda x: x.split(",")[0])
````
